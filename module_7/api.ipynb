## Challenge 
Get (title, full/part time, distance, requirements) from https://app.higherme.com/careers/58bd9e7f472bd?langCode=en when given location and results param

```python
# Imports
import requests as r
```

```python
# Get latitude and longitutde from latlong.net 

location = "Toronto"
def get_lat_long(location):
    url = f"https://geocode.maps.co/search?q={location}"
    resp = r.get(url)
    resp_json = resp.json()
    return resp_json[0]['lat'], resp_json[0]['lon']
# lat, long = 43.653226, -79.3831843 # Replace with function
lat, long = get_lat_long(location)
results = 10
```

```python
url = f"https://api.higherme.com/classic/jobs?page=1&includes=location,location.company&limit={results}&filters[brand.id]=58bd9e7f472bd&filters[lat]={lat}&filters[lng]={long}&filters[distance]=6.25&sort[distance]=asc"

payload = {}
headers = {
  'authority': 'api.higherme.com',
  'accept': 'application/json, text/plain, */*',
  'accept-language': 'en-US,en;q=0.9',
  'higherme-client-version': '2023.07.28_23.1a',
  'origin': 'https://app.higherme.com',
  'sec-ch-ua': '"Not/A)Brand";v="99", "Microsoft Edge";v="115", "Chromium";v="115"',
  'sec-ch-ua-mobile': '?0',
  'sec-ch-ua-platform': '"Windows"',
  'sec-fetch-dest': 'empty',
  'sec-fetch-mode': 'cors',
  'sec-fetch-site': 'same-site',
  'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36 Edg/115.0.1901.188'
}

response = r.get(url, headers=headers, data=payload)

# Lastly, store as a json dict (it will be unsorted)
resp_json = response.json()
resp_json
```

```python
# Now, we parse the data using the json library

final_result = []
for res in resp_json["data"]:
    temp = {}
    temp['title'] = res['attributes']['title']
    if (res['attributes']['full_time'] == True):
        temp['type'] = "Full Time"
    else:
        temp['type'] = "Part Time"
    temp['requirements'] = res['attributes']['requirements']
    temp['distance'] = res['attributes']['distance']
    final_result.append(temp)
final_result
```
